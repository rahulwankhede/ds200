# Plotter

## 1. Bar plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Road_Accidents_2017-Tables_2.5.csv')

x = df['Year'].values
y = df['Number of Accidents - Total'].values

plt.figure(1, figsize=(9,9))
p1 = plt.bar(x, y, 0.8)
plt.ylim(420000, 520000)
plt.yticks(np.arange(420000, 520000, 10000), np.arange(420000, 520000, 10000))
plt.xticks(x, x, rotation=90)
plt.ylabel('Number of total accidents')
plt.title('Total accidents by year')
plt.show()
```

## 2. Box and Whiskers plot

```python
df = pd.read_csv('data/1.12_Trend_of_Production_of_Motor_Vehicles_in_India_1961_1979.csv')

mc = df['Two Wheelers - Motor Cycles'].values
mp = df['Two Wheelers - Mopeds'].values
sc = df['Two Wheelers - Scooters'].values
tw = df['Three Wheelers'].values

data = [mc, mp, sc, tw]

plt.figure(2, figsize=(9,9))
plt.boxplot(data)
plt.xticks([1,2,3,4], ['Motor Cycles', 'Mopeds', 'Scooters', 'Three-wheeler'])
plt.xlabel('Type of Vehicle')
plt.ylabel('Total production of vehicles in India (1961-79) by type')
plt.grid()
plt.show()
```

## 3. Scatter plot

#### Note: I could not find a dataset that would be apt for a scatter plot so I'm genarating my own data randomly and plotting it

```python
np.random.seed(42)

n = 50

x = np.random.uniform(0,10,n)
y = 0.8 * x + 0.3 + np.random.normal(0,1,n)

plt.figure(3, (9,9))
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
plt.grid()
plt.title('Y = 0.8X + 0.3 + epsilon')
```
